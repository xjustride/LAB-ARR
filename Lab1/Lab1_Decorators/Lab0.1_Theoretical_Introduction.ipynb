### 1. Funkcje jako obiekty pierwszej klasy
W Pythonie funkcje traktowane są jako obiekty. Oznacza to, że:
- Mogą być przypisywane do zmiennych.
- Mogą być przekazywane jako argumenty do innych funkcji.
- Mogą być zwracane przez inne funkcje.

Zrozumienie tego faktu jest kluczowe do opanowania techniki dekorowania funkcji.

### 2. Dekoratory
Dekorator to funkcja, która przyjmuje inną funkcję i rozszerza jej zachowanie bez jawnej modyfikacji jej kodu źródłowego.

**Podstawowa składnia:**
```python
import functools

def logowanie(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        print(f"Wywołuję funkcję: {func.__name__}")
        return func(*args, **kwargs)
    return wrapper
```

**Uwaga:** Zastosowanie `@functools.wraps` pozwala na zachowanie metadanych oryginalnej funkcji (takich jak nazwa czy docstring).

### 3. Deskryptory
Deskryptory to fundament mechanizmów takich jak property, classmethod czy staticmethod. Pozwalają one na zdefiniowanie własnej logiki dostępu do atrybutów klasy poprzez metody magiczne: __get__, __set__ oraz __delete__.

### 4. Generatory i Iteratory
Generatory pozwalają na iterowanie po dużych zbiorach danych w sposób leniwy (lazy evaluation), co znacznie oszczędza pamięć operacyjną. Kluczowym słowem jest tutaj yield.

```python
def nieskonczony_licznik():
    n = 0
    while True:
        yield n
        n += 1
```
