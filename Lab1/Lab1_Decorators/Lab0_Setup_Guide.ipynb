## 2. Zarządzanie środowiskami: Conda (Zalecane)
Aby uniknąć konfliktów między bibliotekami, najbezpieczniejszym rozwiązaniem jest użycie **Minicondy**.

### Krok 1: Instalacja
- Pobierz i zainstaluj Minicondę: [https://docs.anaconda.com/miniconda/](https://docs.anaconda.com/miniconda/)

### Krok 2: Tworzenie dedykowanego środowiska
Otwórz terminal i wykonaj następujące polecenia:

```bash
# Tworzenie środowiska o nazwie 'lab_aap' z Pythonem 3.12
conda create -n lab_aap python=3.12 -y

# Aktywacja środowiska
conda activate lab_aap
```

---

## 3. Alternatywa: Lokalny build Pythona (Opcjonalnie)
Możesz skorzystać ze standardowego modułu `venv`, jeśli nie chcesz instalować Condy. Jest to rozwiązanie poprawne, ale w skomplikowanych projektach AI może sprawiać więcej trudności przy zarządzaniu binarnymi zależnościami.

```bash
python -m venv .venv
source .venv/bin/activate  # Mac/Linux
.\.venv\Scripts\activate   # Windows
```

## 4. Instalacja bibliotek i integracja z Jupyter
Po aktywacji środowiska zainstaluj niezbędne pakiety:

```bash
pip install ipykernel notebook
```

Następnie zarejestruj środowisko jako kernel dla Jupyter Notebook:

```bash
python -m ipykernel install --user --name lab_aap --display-name "Python (lab_aap)"
```
